```python
# Q1: Convert the sklearn.dataset cancer to a DataFrame.
#This function should return a (569, 31) DataFrame with
#columns = ...

# method 1: load df_cancer
def answer_one():
    
    # laod dictionary key=cancer
    cancer = load_cancer()
    # ans data structure
    df_cancer = pd.DataFrame()
    
    # saber cuales keys() tiene el diccionario para crear DataFrame
    keys = cancer.keys()
    
    # crear DataFrame pd.DataFrame: Parametros
    # data = cancer.data, cancer.target 
    # columns = cancer.feature_names + key='target' column
    df_cancer = pd.DataFrame(data=np.c_[cancer.data, cancer.target], columns = list(cancer.feature_names) + ['target'] )
    
    # data = cancer['data']
    
    return df_cancer
answer_one()
```

```python
# Q2
# What is the class distribution? (i.e. how many instances of malignant and how many benign?)
# This function should return a Series named target of length 2 with integer values and index = ['malignant', 'benign']


# method 2: count how many malignant and benign entries are in df_cancer
#  (df_cancer['key=target(tumor)']==0/1).sum()
def answer_two():
    # YOUR CODE HERE
    #raise NotImplementedError()
    
    # ans data structure
    target  = pd.Series(name = 'target',index={'malignant', 'benign'})
    
    # load cancer DataFrame 
    df_cancer = answer_one()
    
    mal_tot = (df_cancer['target']==0).sum()
    target['malignant'] = mal_tot
    ben_tot = (df_cancer['target']==1).sum()
    target['benign'] = ben_tot
    
    return target


answer_two()
```

```python
# Q3
#Split the DataFrame into X (the data) and y (the labels).
#This function should return a tuple of length 2: (X, y), where
#X, a pandas DataFrame, has shape (569, 30)
#y, a pandas Series, has shape (569,).

# method 3:  prepare df_cancer for 2 split groups for classifier
def answer_three():
    # YOUR CODE HERE
    #raise NotImplementedError()
    
    # ans data structure
    X, y = pd.DataFrame(),pd.Series()
    
    # load cancer DataFrame
    df_cancer = answer_one()
    
    # save data from DataFrame in X
    X = df_cancer.iloc[:, 0:30]
    
    # save labels from DataFrame in y
    y = df_cancer['target']
    
    # Your code here
    return X,y


answer_three()
```

```python
# Q4
# Using train_test_split, split X and y into 
# training and test sets (X_train, X_test, y_train, and y_test).

# This function should return a tuple of length 4:
# (X_train, X_test, y_train, y_test), where

# X_train has shape (426, 30)
# X_test has shape (143, 30)
# y_train has shape (426,)
# y_test has shape (143,)


# method 4: create train-set and test-set subgroups of classifier

from sklearn.model_selection import train_test_split

def answer_four():
    # YOUR CODE HERE
    
    # ans data structure
    X_train,X_test,y_train,y_test = None,None,None,None
    
    # load prepared data  X-data y labels
    X, y = answer_three()
    
    # create train-set and test-set from prepared data
    
    X_train, X_test, y_train, y_test = train_test_split(X,y,random_state=0)
    
    #print(X_train.shape[0],X_train.shape[1])
    #print(X_test.shape[0],X_test.shape[1])
    #print(y_train.shape[0])
    #print(y_test.shape[0])
    
    
    return X_train, X_test, y_train, y_test

answer_four()
```

```python
# Q5
# Using KNeighborsClassifier, fit a k-nearest neighbors (knn) classifier 
# with X_train, y_train and using one nearest neighbor (n_neighbors = 1).

from sklearn.neighbors import KNeighborsClassifier


# method 5: crear knn nieghbor classifier, con k=1 neighbors

def answer_five():
    # YOUR CODE HERE
    # ans data structure
    knn = KNeighborsClassifier
    
    
    # load train-set and test-set 
    X_train, X_test, y_train, y_test = answer_four()
    
    
    # create knn classifier con k=1 neighbors
    knn = KNeighborsClassifier(n_neighbors = 1)
    
    
    
    
    # fit knn classifier, train-set, entrenar el clasificador
    knn.fit(X_train,y_train)
    
    return knn

answer_five()
```

```python
# Q6
# Using your knn classifier, predict the class label using the mean value for each feature.
# This function should return a numpy array either array([ 0.]) or array([ 1.])

# method 6: load df_cancer, load and fit knn classifier with k=1
# predict output for df_cancer.mean() for all features EXCEPT target

def answer_six():
    # YOUR CODE HERE
    
    # ans data structure
    
    class_label = np.array(np.nan)
    
    # load df_cancer
    df_cancer = answer_one() 
    
    # load and fit/train  knn classifier  k=1
    knn = answer_five()
    
    # get mean value of each feature
    means = df_cancer.mean()[:-1].values.reshape(1, -1)
    
    # predict cancer output
    out_predict = knn.predict( means)
    
    
    return out_predict

answer_six()
```

```python
# Q7
# Using your knn classifier, predict the class labels for the test set X_test.
# This function should return a numpy array with shape (143,) and values either 0.0 or 1.0.

# method 7: load train-set and test-set, load and fit knn class
# predict class labels(features) para el test-set X_test
def answer_seven():
    # YOUR CODE HERE
    
    # ans data structure
    pred_class_labl = np.zeros(143,)
    
    # load train-set and test-set
    X_train, X_test, y_train, y_test = answer_four()
    
    # load and fit knn classifier
    knn = answer_five()
    
    # predict class labels (features) para el test-set X_test
    pred_class_labl = knn.predict(X_test)
    #print(pred_class_labl.shape[0])
    #print(pred_class_labl[10,] )
    
    return pred_class_labl
answer_seven()
```

```python
# Q8
# Find the score (mean accuracy) of your knn 
# classifier using X_test and y_test.
# This function should return a float between 0 and 1


def answer_eight():
    # YOUR CODE HERE
    
    # ans data structure
    performance = np.nan
    
    # load train-set and test-set
    X_train, X_test, y_train, y_test = answer_four()
    
    # load and fit knn classifier, with k=5
    knn = answer_five()
    
    # score/measure performance of classifier with test-set X_test and y_test
    performance = knn.score(X_test,y_test)
    
    return performance
answer_eight()
```

```python
# visualize the differet predicition scores 
# between training and test sets, as well as malignant and benign cells.

def accuracy_plot():
    import matplotlib.pyplot as plt

    %matplotlib notebook

    X_train, X_test, y_train, y_test = answer_four()

    # Find the training and testing accuracies by target value (i.e. malignant, benign)
    mal_train_X = X_train[y_train==0]
    mal_train_y = y_train[y_train==0]
    ben_train_X = X_train[y_train==1]
    ben_train_y = y_train[y_train==1]

    mal_test_X = X_test[y_test==0]
    mal_test_y = y_test[y_test==0]
    ben_test_X = X_test[y_test==1]
    ben_test_y = y_test[y_test==1]

    knn = answer_five()

    scores = [knn.score(mal_train_X, mal_train_y), knn.score(ben_train_X, ben_train_y), 
              knn.score(mal_test_X, mal_test_y), knn.score(ben_test_X, ben_test_y)]


    plt.figure()

    # Plot the scores as a bar chart
    bars = plt.bar(np.arange(4), scores, color=['#4c72b0','#4c72b0','#55a868','#55a868'])

    # directly label the score onto the bars
    for bar in bars:
        height = bar.get_height()
        plt.gca().text(bar.get_x() + bar.get_width()/2, height*.90, '{0:.{1}f}'.format(height, 2), 
                     ha='center', color='w', fontsize=11)

    # remove all the ticks (both axes), and tick labels on the Y axis
    plt.tick_params(top='off', bottom='off', left='off', right='off', labelleft='off', labelbottom='on')

    # remove the frame of the chart
    for spine in plt.gca().spines.values():
        spine.set_visible(False)

    plt.xticks([0,1,2,3], ['Malignant\nTraining', 'Benign\nTraining', 'Malignant\nTest', 'Benign\nTest'], alpha=0.8);
    plt.title('Training and Test Accuracies for Malignant and Benign Cells', alpha=0.8)
    
accuracy_plot()
```
