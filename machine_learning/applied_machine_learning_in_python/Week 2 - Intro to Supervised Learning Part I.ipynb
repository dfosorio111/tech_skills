```python
# General Concepts Intro:

# X  data: sample features 
# y target value: target label(output) to predict
# X_train, y_train 75%  train-set,  X_test, y_test 25% test-set
# knn.fit(X_train,y_train) train-set, entrena el modelo para estimar parametros!!
# knn.fit(X_train,y_train) obtiene parametros del modelo!!!
# knn.score(X_test,y_test) test_set  medir performace de modelo
# knn.predict(object vals(X data new features))  predecir y target label dadas caracteristicas o parametros data X
```

```python
# Types of Supervised Learning
# Supervised Learning: Classification and Regression
# take train-set y map en target value

# Classification:  y target value es discreto, 
# Muti-class Classification: y target es un set de values
# Multi-label Classification: multiple y target 


# Regression: y target value es continuo

# entre más complexity parameters del modelo, más accuracy
```

```python
# Week 2 - Overfitting and Underfitting
# underfitting: too simple model even for train-set
# overfitting: too complex model, too sensitive to changes and not good to generalize y target
```

```python
# classification: find desicion boundary,  binary y target 

# knn: disminuir el k aumenta la varianza de las desicion boundaries
```

```python
# pd.read_table  load fuits DataSet from .txt file
fruits = pd.read_table('fruit_data_with_colors.txt')

# over-write columns to create X features  
X_fruits = fruits[['height', 'width', 'mass', 'color_score']]
# over-write unique colum of y target
y_fruits = fruits['fruit_label']
# list de classification  y target
target_names_fruits = ['apple', 'mandarin', 'orange', 'lemon']

# over-write X features and y target for plotting
X_fruits_2d = fruits[['height', 'width']]
y_fruits_2d = fruits['fruit_label']

# train_test_split(X_train,y_train,random_state=0(75-25%) )
# divide DataSet(Frame) en train-set y test-set
X_train, X_test, y_train, y_test = train_test_split(X_fruits, y_fruits, random_state=0)

from sklearn.preprocessing import MinMaxScaler
scaler = MinMaxScaler()

X_train_scaled = scaler.fit_transform(X_train)
# we must apply the scaling to the test set that we computed for the training set
X_test_scaled = scaler.transform(X_test)

# crear modelo knn con k=n_neighbors=5
knn = KNeighborsClassifier(n_neighbors = 5)
# knn.fit(X_train,y_train) train  train-set, entrena el modelo para estimar parametros!!
knn.fit(X_train_scaled, y_train)
# knn.score(X,y) medir performance del modelo con los parametros obtenidos de training
print('Accuracy of K-NN classifier on training set: {:.2f}'
     .format(knn.score(X_train_scaled, y_train)))
print('Accuracy of K-NN classifier on test set: {:.2f}'
     .format(knn.score(X_test_scaled, y_test)))
# knn.predict(object X features)
example_fruit = [[5.5, 2.2, 10, 0.70]]
example_fruit_scaled = scaler.transform(example_fruit)
print('Predicted fruit type for ', example_fruit, ' is ', 
          target_names_fruits[knn.predict(example_fruit_scaled)[0]-1])
```

```python
from sklearn.datasets import make_classification, make_blobs
from matplotlib.colors import ListedColormap
from sklearn.datasets import load_breast_cancer
from adspy_shared_utilities import load_crime_dataset

cmap_bold = ListedColormap(['#FFFF00', '#00FF00', '#0000FF','#000000'])


# synthetic dataset for simple regression
from sklearn.datasets import make_regression
plt.figure()
plt.title('Sample regression problem with one input variable')
# make_regression(n_samples, n_features, n_informative, bias, noise, random_state)
# create regression problem con # samples, # features,...
X_R1, y_R1 = make_regression(n_samples = 100, n_features=1,
                            n_informative=1, bias = 150.0,
                            noise = 30, random_state=0)
# make scatter (discrete) plot 
plt.scatter(X_R1, y_R1, marker= 'o', s=50)
plt.show()
```

```python
# synthetic dataset for more complex regression
from sklearn.datasets import make_friedman1
plt.figure()
# regression model con 1 input variable (X feature)
plt.title('Complex regression problem with one input variable')
# make_friedman1(n_samples, n_features, random_state=0)
X_F1, y_F1 = make_friedman1(n_samples = 100,
                           n_features = 7, random_state=0)

plt.scatter(X_F1[:, 2], y_F1, marker= 'o', s=50)
plt.show()

# synthetic dataset for classification (binary) 
plt.figure()
plt.title('Sample binary classification problem with two informative features')
# binary classification: 2 class datasets con 2 X features
# make_classification(n_samples, n_features,
                                #n_redundant, n_informative,
                                #n_clusters_per_class, flip_y,
                                #class_sep, random_state=0)

X_C2, y_C2 = make_classification(n_samples = 100, n_features=2,
                                n_redundant=0, n_informative=2,
                                n_clusters_per_class=1, flip_y = 0.1,
                                class_sep = 0.5, random_state=0)
plt.scatter(X_C2[:, 0], X_C2[:, 1], c=y_C2,
           marker= 'o', s=50, cmap=cmap_bold)
plt.show()
```

```python
# more difficult synthetic dataset for classification (binary) 
# with classes that are not linearly separable

# Complex binary classification dataset, con 2  X features no linealmente independientes
# make_blobs(n_samples,n_features,centers, cluster_std, random_state)

X_D2, y_D2 = make_blobs(n_samples = 100, n_features = 2, centers = 8,
                       cluster_std = 1.3, random_state = 4)
y_D2 = y_D2 % 2
plt.figure()
plt.title('Sample binary classification problem with non-linearly separable classes')
plt.scatter(X_D2[:,0], X_D2[:,1], c=y_D2,
           marker= 'o', s=50, cmap=cmap_bold)
plt.show()
```
