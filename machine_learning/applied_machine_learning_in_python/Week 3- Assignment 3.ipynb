```python
# method 1: get percentage/ratio of parameter=class 'Class' fraud, binary classification 0-1
def answer_one():
    
    # Your code here
    
    # ans data structure
    percen_fraud = float()
    
    # load source data
    source_data = pd.read_csv('fraud_data.csv')
    
    # over-over write source_data DataFrame, with class 'Class'==1
    data_fraud = source_data[source_data['Class']==1]
    
    # get percentage of fraud, from total entries in DataFrame
    percen_fraud = len(data_fraud)/len(source_data)
    
    #print(type(percen_fraud))
    return percen_fraud# Return your answer
answer_one()
```

```python
# Use X_train, X_test, y_train, y_test for all of the following questions
from sklearn.model_selection import train_test_split

df = pd.read_csv('fraud_data.csv')

X = df.iloc[:,:-1]
y = df.iloc[:,-1]

X_train, X_test, y_train, y_test = train_test_split(X, y, random_state=0)
```

```python
# method 2: create Dummy Classifier with strategy=most_frequent(majority class), get evaluation metrics
def answer_two():
    from sklearn.dummy import DummyClassifier
    from sklearn.metrics import recall_score
    
    # Your code here
    
    # ans data structure
    eval_metrics = ()
    
    # create a DummyClassifier, parameters: strategy = 'most_frequent', fit/train with train-set
    dummy_majority = DummyClassifier(strategy = 'most_frequent').fit(X_train, y_train)
    
    # get predicted values from DummyClassifier
    y_predicted = dummy_majority.predict(X_test)
    
    # score/performance of DummyClassifier, accuracy evaluation metric of model
    score = dummy_majority.score(X_test,y_test)
    
    # get accuracy score of model, using y_test and y_predicted
    acc = accuracy_score(y_test, y_predicted)
    
    # get recall of model, using y_test and y_predicted
    recall = recall_score(y_test, y_predicted)
    
    eval_metrics = score, recall
    
    #print(type(score), type(recall))
    #print(type(eval_metrics))
    return eval_metrics# Return your answer

answer_two()
```

```python
# method 3: create SVectorMachine model, get evaluation metrics
def answer_three():
    from sklearn.metrics import recall_score, precision_score
    from sklearn.svm import SVC
    
    # Your code here
    
    # ans data structure
    eval_metrics = ()
    
    # create SVC_model, parameters: random_state=0, fit/train
    SVC_model = SVC(random_state=0).fit(X_train,y_train)
    
    # get predicted values from the model using y_test
    y_predicted = SVC_model.predict(X_test)
    
    # get confusion matrix for the model with y_test and y_predicted
    confusion = confusion_matrix(y_test,y_predicted)
    #print(confusion)
    
    # score/accuracy measure performance of model with test-set
    score = SVC_model.score(X_test,y_test)
    
    # get evaluation metrics for the model, using y_test and y_predicted
    acc = accuracy_score(y_test, y_predicted)
    recall = recall_score(y_test, y_predicted)
    precision = precision_score(y_test, y_predicted)
    
    eval_metrics = (acc, recall, precision)
    
    
    #print(type(acc), type(recall), type(precision))
    #print(type(eval_metrics))
    return eval_metrics# Return your answer

answer_three()
```

```python
def answer_six():    
    from sklearn.model_selection import GridSearchCV
    from sklearn.linear_model import LogisticRegression

    # Your code here
    
    # ans data structure
    mean_test_scores =np.zeros((5,2))
    
    # create LogisticClassifier, default parameters, fit/train with train-set
    logistic_reg = LogisticRegression().fit(X_train,y_train)
    
    # grid value parameters to iterate over, optimization of parameters
    grid_values_param = {'penalty':['l1','l2'], 'C':[0.01, 0.1, 1, 10, 100]}
    
    # run GridSearchCV, paramters: model=logistic_reg, param_grid = grid_values_param,  scoring='recall'
    #  cross-validation= default 3-fold  cv=3 (None)
    # fit/train GridSearchCV with train-set
    grid_log_recall = GridSearchCV(logistic_reg, param_grid = grid_values_param, scoring = 'recall').fit(X_train,y_train)
    
    # get mean test scores of GridSearch from combination of parameter values
    mean_test_scores = np.array([grid_log_recall.cv_results_['mean_test_score'][x:x+2] for x in range(0, len(grid_log_recall.cv_results_['mean_test_score']), 2)])
    
    return mean_test_scores# Return your answer

answer_six()
```
