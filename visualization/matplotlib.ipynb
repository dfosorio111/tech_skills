## Scatter Plots: Gráfica discreta

```python

# librerias
import numpy as np

# array de datos x,y Now let's create some fake data
x = np.array([1,2,3,4,5,6,7,8])
y = x

# crear Figure

plt.figure()

# And now let's get matplotlib to produce a scatter plot. This is similar to plt.plot(x, y, '.'), but
# crear scatter discreta con puntos de vectores x,y  
plt.scatter(x, y) 
```

```python
# vectores x,y
x = np.array([1,2,3,4,5,6,7,8])
y = x

# lista de colores
# ['green', 'green', 'green', 'green', 'green', 'green', 'green', 'red']
colors = ['green']*(len(x)-1)
colors.append('red')

# crear Figure
plt.figure()

# plot the point with size 100 and our chosen color list
# plot scatter plot con puntos x,y tamaño s, colores de lista 
plt.scatter(x, y, s=100, c=colors)
```

```python
plt.figure()

# plot scatter plot con vectores x,y sliced, color c='red' 
plt.scatter(x[:2], y[:2], s=100, c='red', label='Tall students')
# plot scatter plot con vectores x,y sliced, color c='blue'
plt.scatter(x[2:], y[2:], s=100, c='blue', label='Short students');

# label x
plt.xlabel('The number of times the child kicked a ball')
# label y
plt.ylabel('The grade of the student')
# titulo de Figura
plt.title('Relationship between ball kicking and grades')
# convencion/leyenda de cada plot
plt.legend(loc=4, frameon=False, title='Leyenda')
```

## Lineplot: Gráfica continua

```python
# datos x,y para lineplot
linear_data = np.array([1,2,3,4,5,6,7,8])
exponential_data = linear_data**2

# crear Figura
plt.figure()

# plot lineplot(datos1,'form',datos2,'form')     
plt.plot(linear_data, '-o', exponential_data, '-o')

# crear leyenda
plt.legend(['Competition', 'Us'])

# funcion fill_between entre graficas
# fill_between(range(len(datos), datos, facecolor='color', alpha=alpha)
plt.gca().fill_between(range(len(linear_data)), 
                       linear_data, exponential_data, 
                       facecolor='blue', 
                       alpha=0.25)
```

```python
# datos con fechas
observation_dates = np.arange('2017-01-01', '2017-01-09', dtype='datetime64[D]')

# line plot plot(x,y,'tipo')
plt.plot(observation_dates, linear_data, '-o',  observation_dates, exponential_data, '-o')

# variable x axis
x = plt.gca().xaxis

# rotar items de xaxis
for item in x.get_ticklabels():
    item.set_rotation(45)
```

```python
# datos
observation_dates = np.arange('2017-01-01', '2017-01-09', dtype='datetime64[D]')
# line plot plot(x,y,'tipo')
plt.plot(observation_dates, linear_data, '-o',  observation_dates, exponential_data, '-o');

x = plt.gca().xaxis # variable xaxis

# rotar items de xaxis
for item in x.get_ticklabels():
    item.set_rotation(45)
    

# cambiar labels
ax = plt.gca()
ax.set_xlabel('Date') # xaxis label
ax.set_ylabel('Units') # yaxis label
ax.set_title('Exponential vs. Linear performance'); # crear titulo
```

```python
plt.figure(figsize=(8,6))

# datos x
observation_dates = np.arange('2017-01-01', '2017-01-09', dtype='datetime64[D]')

plt.plot(observation_dates, linear_data, '-o',  observation_dates, exponential_data, '-o'); # plot line plot
x = plt.gca().xaxis

# rotar items de xaxis
for item in x.get_ticklabels():
    item.set_rotation(45)
    
# labels
ax = plt.gca()
ax.set_xlabel('Date')
ax.set_ylabel('Units')

# titulo
ax.set_title("Exponential ($x^2$) vs. Linear ($x$) performance");
```

## Bar chart: Gráfico de barras

```python
# datos x
linear_data = np.array([1,2,3,4,5,6,7,8])

# lista de x label 
xvals = range(len(linear_data))
# And send this along with the linear data's bar heights, to the bar function. 
# We pass in the width of the bar as 0.3.

# bar chart bar(x,y,width=w )  
plt.bar(xvals, linear_data, width = 0.3)
```

```python
# vector de nuevos datos
new_xvals = []
exponential_data = linear_data**2

# agregar item a valores de x
for item in xvals:
    new_xvals.append(item+0.3)

# plot bar chart  bar(x, y, width = w)
plt.bar(xvals, linear_data, width = 0.3)
plt.bar(new_xvals, exponential_data, width = 0.3 ,color='red')
```

```python
from random import randint
# vector de error de datos y 
linear_err = [randint(1,4) for x in range(len(linear_data))] 

# bar chart bar(x, y, width = w, yerr= vector_error)
plt.bar(xvals, linear_data, width = 0.3, yerr=linear_err)
```

```python
# valores de x
xvals = range(len(linear_data))
# bar chart bar(x,y,width=w,color='c')
plt.bar(xvals, linear_data, width = 0.3, color='b')
plt.bar(xvals, exponential_data, width = 0.3, bottom=linear_data, color='r')
```
